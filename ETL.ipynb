# Transform Dataset

```python
df = df.dropna(subset=["STARS"])
df
```

```python
import re
def findDirector(x):
  result = re.search('Director:\n(.*)\n| \n Star', x)
  if result==None:
    return "Anonymous Director"
  else:
    return result.group(1)
def findStars(x):
  result = re.search('Stars:\n(.*), \n(.*), \n(.*), \n(.*)\n', x)
  if result==None:
    return "Anonymous Stars"
  else:
    return ', '.join(list(result.groups()))
```

```python
df["Director"] = df["STARS"].apply(findDirector)
df["Stars"] = df["STARS"].apply(findStars)
```

```python
df = df.drop(['ONE-LINE','STARS','VOTES','RunTime'],axis=1)
df
```

```python
def findGenre(x):
  result = re.search('\n(.*)', x)
  if result==None:
    return "Anonymous Genre"
  else:
    return result.group(1)
```

```python
df["GENRE"] = df["GENRE"].apply(str)
df["GENRE"] = df["GENRE"].apply(findGenre)
```

```python
df
```

```python
p = df["Director"].unique()
```

```python
ip = []
for i in range(len(p)):
  id = 'Dir('+str(i+1)+')'
  ip.append(id)
```

```python
m = df["MOVIES"].unique()
im = []
for i in range(len(m)):
  id = 'M('+str(i+1)+')'
  im.append(id)
```

```python
id_film = {m[i]: im[i] for i in range(len(m))}
```

```python
def movieid(x):
  return id_film[x]
```

```python
df["ID Movie"] = df["MOVIES"].apply(movieid)
```

```python
id_director = {p[i]:ip[i] for i in range(len(p))}
```

```python
def directorid(x):
  return id_director[x]
```

```python
df["ID Director"] = df["Director"].apply(directorid)
```
