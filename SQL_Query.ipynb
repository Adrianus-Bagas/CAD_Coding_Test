```python
import pandas as pd
from pandasql import sqldf
```

```python
movie_table = pd.read_csv('/content/drive/MyDrive/dataset/movie_table.csv')
director_table = pd.read_csv('/content/drive/MyDrive/dataset/director_table.csv')
```

```python
data = sqldf('''select movie_table.ID_Movie, movie_table.Movie_Name, 
movie_table.Year, movie_table.Genre, movie_table.Rating, 
movie_table.Stars, movie_table.Gross, 
director_table.Director_Name from movie_table, director_table where movie_table.ID_Director=director_table.ID_Director''')
data
```

# Number of unique film titles

```python
sqldf('select count(distinct Movie_Name) as Number_of_Films from data')
```

# Film Title, Year of Release, and Rating of the film starring Lena Headey Sort By Year of Release

```python
sqldf('select Movie_Name, Year, Rating from data where Stars like "%Lena Headey%" order by Year')
```

# The name of the director and total gross of the films that have been directed

```python
sqldf('select Director_Name, Gross as Total_Gross from data group by Director_Name')
```

# Film Title, Year of Release, and Rating of 5 films that have comedy genre with the largest gross

```python
sqldf('select distinct Movie_Name, Year, Rating from data where Genre like "%Comedy%" order by Gross desc limit 5')
```

# Film Title, Year of Release and Rating of the film directed by Martin Scorsese and starring Robert De Niro

```python
sqldf('select Movie_Name, Year, Rating from data where Director_Name="Martin Scorsese" and Stars like "%Robert De Niro%"')
```
